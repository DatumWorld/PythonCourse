## pandas输入和输出（一）

### 比较常用的读写 csv excel 数据库

```python
from faker import Faker
fake = Faker('zh_CN')
```

```python
import pandas as pd
df=pd.DataFrame(data=[ [fake.name(),fake.company(),fake.numerify(),'空格 空格2 空格3'] for x in range(10) ],columns=['user','company','number','空格列'])
```

```python
df #生成的测试数据来演示
```

```python
df.to_csv('demo.csv',index=False) # 保存成csv
```

```python
pd.read_csv('demo.csv') # 读取刚才保存的csv数据文件。
# 注意现在有个索引列，最左边的那列。并不需要.    通过刚才的调整，保存csv时没有保存索引列（index=False）
```

```python
df.to_excel('demo.xlsx',index=False) # 保存成excel
```

```python
pd.read_excel('demo.xlsx') # 读取刚刚保存的excel
# 类似保存csv时，去掉了索引列（index=False）
```
