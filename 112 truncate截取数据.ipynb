## 112 truncate截取数据

### truncate ：根据条件按照索引值截取数据，去掉before之前和after之后的数据。对照loc
参数 before=None, after=None, axis=None, copy=True

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='3H')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[1,3]=np.nan #  修改成nan
```

### 演示数据

```python
df 
```

```python
df.set_index('时间').loc['2020-01-01':'2020-01-02'] # 注意'2020-01-02'，说明日期部分匹配就可以
```

```python
df.set_index('时间').truncate(before='2020-01-01',after='2020-01-02') # 提示：只能作用于有序索引
# 2020-01-02 == 2020-01-02 00:00:00
```

```python
df.truncate(before=2,after=6) # 提示：只能作用于有序索引
```

```python
df.truncate(2,6) # “掐掉”2之前和6 之后的 行，留下中间的
```
