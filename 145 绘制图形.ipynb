## 145 绘制图形

### plot：绘制图形。含bar
##### 绘制图形功能不是pandas的重点，仅作简单介绍，pandas默认使用matplotlib作为图形后端（pip install matplotlib 安装）。
##### 绘图推荐使用专用的绘图工具如matplotlib等

```python

import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[1,1]=np.nan #  修改成nan
```

```python
from pylab import mpl
# 图片文字设置
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
df
```

```python
df.plot.bar(x='时间',y='销售额',xlabel='时间1',ylabel='销售',figsize=(10,5)) # 简单的例子，bar图。同样效果，与下面对照
```

```python
df.plot(x='时间',y='销售额',xlabel='时间1',ylabel='销售',figsize=(10,5),kind='bar') # 简单的例子，bar图
```

```python
df.plot(x='时间',y='销售额',xlabel='时间1',ylabel='销售',figsize=(10,5),kind='line') # 简单的例子，折线图
#(x='x轴的数据',y='y轴的数据',xlabel='x轴的标签',ylabel='y轴的标签',figsize=图形大小,kind='图表的类型'
```

```python
df.plot(x='时间',y='销售额') # 简单的例子，折线图
```
