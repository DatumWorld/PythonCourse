## 170 读写excel更多操作

###   to_excel read_excel 更多操作：表名称 编码 多表格

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[1,1]=np.nan #  修改成nan
```

```python
df
```

```python
with pd.ExcelWriter('demo_multi.xlsx') as writer:
    df.to_excel(writer,sheet_name='table1',index=False,encoding='gbk')
    df[:5].copy().to_excel(writer,sheet_name='table2',index=False,encoding='gbk')
```

```python
pd.read_excel('demo_multi.xlsx',sheet_name=['table2','table1'])['table2'] # 依次读取多个表，并获取某个名字的表
```

```python
pd.read_excel('demo_multi.xlsx',sheet_name=['table2','table1']).keys() # # 依次读取多个表的表格名称
```

```python
pd.read_excel('demo_multi.xlsx') # 默认读取第一个位置的表
```

```python
df.to_excel('demo.xlsx',sheet_name='demo',encoding='gbk',index=False,engine='openpyxl') # sheet_name 保存的表格的名称 ；encoding 编码  ，index 是否保存索引
```

```python
pd.read_excel('demo.xlsx',sheet_name='demo',index_col=False) # index_col 设置索引
```

```python
pd.read_excel('demo.xlsx',sheet_name='demo',index_col='dept') # index_col 设置索引
```

```python
pd.read_excel('demo.xlsx',sheet_name='demo',index_col=[0,1]) # index_col 设置索引
```
