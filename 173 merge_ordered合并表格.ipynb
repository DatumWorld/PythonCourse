## 173 merge_ordered合并表格

### merge_ordered合并表格：为有序数据设计，可以分组或填充nan
left, right, on=None, left_on=None, right_on=None, left_by=None, right_by=None, fill_method=None, suffixes=('_x', '_y'), how='outer'

```python
import numpy  as np
import pandas as pd
```

```python
df1 = pd.DataFrame(
    {
        "key": ["a", "c", "e", "a", "c", "e"],
        "lvalue": [1, 2, 3, 1, 2, 3],
        "group": ["a", "a", "a", "b", "b", "b"]
    }
)
df1
```

```python
df2 = pd.DataFrame({"key": ["b", "c", "d"], "rvalue": [4, 5, 6]})
df2
```

```python
pd.merge_ordered(df1,df2,left_by='group',fill_method='ffill',how='left') # how 表的联结方式
```

```python
pd.merge_ordered(df1,df2,left_by='group',fill_method='ffill') # fill_method 填充nan
```

```python
pd.merge_ordered(df1,df2,left_by='group') # group 分组后合并
```

```python
pd.merge_ordered(df1,df2) # 对照merge
```
