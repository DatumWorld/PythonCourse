##  pandas表格合并

### 表格合并
#### join merge 

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
```

```python
df.insert(0,'dept',['部门1']*5+['部门2']*5)
```

```python
df # 随机生成的数据
```

```python
df1=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name1'])
df1['销售额']=np.random.randint(100,100000,size=(10, 1))
df1.insert(0,'dept',['部门1']*5+['部门2']*5)
```

```python
df1  # 随机生成的数据
```

```python
df.join(df1,lsuffix='_l',rsuffix='_r') # 两个表格合并，如果列名相同需要指定 后缀
```

```python
df.merge(df1,how='left',left_on='dept',right_on='dept') # 两个表格合并，类似SQL 语句 的合并方式，可以指定 连接方式how ，连接on的列名（left_on 和 right_on）
```
