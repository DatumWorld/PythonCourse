### 分组聚合统计
#### groupby 多级索引 first last sort

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['奖金']=np.random.randint(100,100000,size=(10, 1))
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df # 随机生成的数据
```

```python
df.groupby(['name'],sort=False).last() # 分组的列不排序，按照原来的顺序。默认（如下面）会排序
```

```python
df.groupby(['name']).last() # 分组后取各组最后一个
```

```python
df.groupby(['dept','dept1']).last() # 分组后取各组最后一个
```

```python
groups=df.groupby(['dept','dept1'])
```

```python
df[::-1].groupby(['dept','dept1']).first() # 验证效果，将df 的行顺序反过来分组取第一个
```

```python
groups.first() # 分组后取各组第一个
```

```python
groups.groups
```
