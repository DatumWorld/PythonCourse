### 数据抽样进阶
#### sample 上采样 抽取可重复结果 抽取固定 按照权重抽样

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['姓名'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['工龄']=np.random.randint(1,10,size=(10, 1))
df.insert(0,'部门',['部门1']*5+['部门2']*5)
df['绩效系数']=np.random.randint(7,10,size=(10, 1))/10
df['补贴']=np.random.randint(10,1000,size=(10, 1))
df['请假扣除']=np.random.randint(0,1000,size=(10, 1))
df['权重']=([0.2]+[.0])*5
```

```python
df # 随机生成的数据
```

```python
df.sample(frac=.6,weights='权重',replace=1) # 按照权重列的权重采样。上采样。需要结果中可以重复出现数据。注意权重大于零的有五个，所以需要考虑是否允许重复数据
```

```python
df.sample(frac=2,weights='权重',replace=1) # 按照权重列的权重采样。上采样。需要结果中可以重复出现数据
```

```python
df.sample(frac=2,replace=1) # 上采样。需要结果中可以重复出现数据
```

```python
df.sample(n=3,replace=0) # 一次结果中不可以出现重复数据
```

```python
df.sample(n=3,replace=1) # 结果中可以出现重复数据
```

```python
df.sample(n=3,random_state=1) # 抽样结果不变
```

```python
df.sample(n=3) # 普通抽样
```
