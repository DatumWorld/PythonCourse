## 53 pandas透视表

### 类似excel透视表

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['姓名'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['工龄']=np.random.randint(1,10,size=(10, 1))
df.insert(0,'部门',['部门1']*5+['部门2']*5)
df['绩效系数']=np.random.randint(7,10,size=(10, 1))/10
df['补贴']=np.random.randint(10,1000,size=(10, 1))
df['请假扣除']=np.random.randint(0,1000,size=(10, 1))
df['权重']=([0.2]+[.0])*5
```

```python
df # 随机生成的数据
```

```python
pd.pivot_table(df,values='销售额',index='部门',columns=['姓名'],aggfunc=[np.sum,np.mean]) # 另一种等效写法
```

```python
pd.pivot_table(df,values='销售额',index='部门',columns='姓名',aggfunc=['sum','mean']) # 计算多个统计指标
```

```python
pd.pivot_table(df,index='部门',columns='姓名',aggfunc=sum) # 这时候是对多个列计算
```

```python
pd.pivot_table(df,values='销售额',index='部门',columns='姓名',aggfunc=sum) # 一个简单的透视表
```
