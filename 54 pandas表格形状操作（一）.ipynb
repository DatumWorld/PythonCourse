### pivot，类似pivot_table的某列转列头的操作，但是不支持重复标签对（指行/列对）

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['姓名'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['工龄']=np.random.randint(1,10,size=(10, 1))
df.insert(0,'部门',['部门1']*5+['部门2']*5)
df['绩效系数']=np.random.randint(7,10,size=(10, 1))/10
df['补贴']=np.random.randint(10,1000,size=(10, 1))
df['请假扣除']=np.random.randint(0,1000,size=(10, 1))
df['权重']=([0.2]+[.0])*5
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df # 随机生成的数据
```

```python
df.pivot(values=['销售额','补贴'],index=['部门','dept1'],columns=['姓名','权重']) # 以部门 + dept1 列为index，姓名 + 权重 列 转为列标题，销售额 补贴 作为值调整表格。
# 无意义，仅展示用法
```

```python
df.pivot(values=['销售额','补贴'],index=['部门','dept1'],columns='姓名')['销售额']  # 以部门 + dept1 列为index，姓名列 转为列标题，销售额 补贴 作为值调整表格。
# 获取 销售额 的值，去掉其他的
```

```python
df.pivot(values=['销售额','补贴'],index=['部门','dept1'],columns='姓名')  # 以部门 + dept1 列为index，姓名列 转为列标题，销售额 补贴 作为值调整表格。
```

```python
df.pivot(values='销售额',index=['部门','dept1'],columns='姓名')  # 以部门 + dept1 列为index，姓名列 转为列标题，销售额作为值调整表格
```

```python
df.pivot(values='销售额',index='部门',columns='姓名')  # 以部门列为index，姓名列 转为列标题，销售额作为值调整表格
```
