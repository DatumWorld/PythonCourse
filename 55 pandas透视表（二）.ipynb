## 55 pandas透视表(二)

### 类似excel透视表，进阶。空值/缺失值/合计列等

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['姓名'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['工龄']=np.random.randint(1,10,size=(10, 1))
df.insert(0,'部门',['部门1']*5+['部门2']*5)
df['绩效系数']=np.random.randint(7,10,size=(10, 1))/10
df['补贴']=np.random.randint(10,1000,size=(10, 1))
df['请假扣除']=np.random.randint(0,1000,size=(10, 1))
df['权重']=([0.2]+[.0])*5
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df.loc[0,'销售额']=np.nan
```

```python
df # 随机生成的数据
```

```python
df.pivot_table(values={'销售额':[np.sum,np.mean],'补贴':[np.sum,np.mean]},index=['部门','dept1'],columns=['姓名']) # 可以指定多个列的指标
```

```python
df.pivot_table(values={'销售额':np.sum},index=['部门','dept1'],columns=['姓名']) # 可以利用字典dict自定义列
```

```python
df.pivot_table(values=['销售额'],columns=['dept1'],index=['部门'],fill_value='-',margins=1,margins_name='合计',aggfunc=sum) # 新增：添加合计行/列。设置合计列/行的名字
```

```python
df.pivot_table(values=['销售额'],columns=['dept1'],index=['部门'],fill_value='-',margins=1,margins_name='合计/平均',aggfunc=[sum,np.mean]) # 新增：奇特写法。与上面对照
```

```python
df.pivot_table(values=['销售额'],columns=['dept1'],index=['部门'],fill_value='-',margins=1,margins_name='平均') # 添加合计行/列。设置合计列/行的名字。默认平均
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名'],index=['部门','dept1'],fill_value='-',margins=1,margins_name='合计',aggfunc=sum) # 添加合计行/列。设置合计列/行的名字
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名'],index=['部门','dept1'],fill_value='-',margins=1,margins_name='平均',aggfunc=np.mean) # 添加合计行/列。设置合计列/行的名字
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名'],index=['部门','dept1'],fill_value=0,margins=1) # 添加合计行/列
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名'],index=['部门','dept1'],dropna=0) # 是否删除na
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名'],index=['部门','dept1'],fill_value=0) # 将结果的na 使用0填充
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名'],index=['部门','dept1']) # values columns index 可以多列
```

```python
df.pivot_table(values=['销售额','补贴'],columns=['姓名','工龄'],index=['部门','dept1']) # values columns index 可以多列。无意义的演示
```
