## 57 pandas交叉表 

### crosstab 计算因素的简单的交叉表

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['姓名'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['工龄']=np.random.randint(1,10,size=(10, 1))
df.insert(0,'部门',['部门1']*5+['部门2']*5)
df['绩效系数']=np.random.randint(7,10,size=(10, 1))/10
df['补贴']=np.random.randint(10,1000,size=(10, 1))
df['请假扣除']=np.random.randint(0,1000,size=(10, 1))
df['权重']=([0.2]+[.0])*5
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df # 随机生成的数据
```

```python
pd.crosstab(index=df.部门,columns=df.dept1,rownames=['大部门'],colnames=['小部门'],dropna=0,margins=True,margins_name='合计',normalize='all') 
# 进行占比计算。可以分别按照 行/列 和总体进行计算
```

```python
pd.crosstab(index=df.部门,columns=df.dept1,rownames=['大部门'],colnames=['小部门'],dropna=0,margins=True,margins_name='合计',normalize=True) 
# 进行占比计算
```

```python
pd.crosstab(index=df.部门,columns=df.dept1,rownames=['大部门'],colnames=['小部门'],dropna=0,margins=True,margins_name='合计') 
# 加上合计 列/行。自定义 合计 行/列 名字
```

```python
pd.crosstab(index=df.部门,columns=df.dept1,rownames=['大部门'],colnames=['小部门'],dropna=0,margins=True) 
# 加上合计 列/行
```

```python
pd.crosstab(index=df.部门,columns=df.dept1,rownames=['大部门'],colnames=['小部门'],dropna=0) 
# dropna 是否删除 均是nan 的 列
```

```python
pd.crosstab(index=[df.部门,df.dept1],columns=['销售额'],values=df.销售额 ,aggfunc=sum, rownames=['大部门','小部门'],colnames=['销售额']) 
# 类似 groupby 
```

```python
pd.crosstab(df.部门,df.dept1,rownames=['大部门'],colnames=['小部门']) # 一个简单的交叉表，大部门含有 小部门 的个数。指定列名/行头
```

```python
pd.crosstab(df.部门,df.dept1) # 一个简单的交叉表，部门含有 dept1 的个数
```
