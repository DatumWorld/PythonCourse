## 81 assign计算更新列

### assign ：新增列或修改已有列

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

### 演示数据

```python
df.iloc[0,3]=np.nan #  修改成nan
df.iloc[1,3]=df.iloc[5,3] #  修改成一样
```

```python
df # 随机生成的数据
```

```python
df.assign(new=df.销售额*2,
         new_1 =df.时间,
         新_3=lambda x:x.new+10000,
         新_4=lambda x:x.dept+x['dept 1'],
         新_5=lambda x:x.dept+x['dept 1']+x.新_3.astype(str),
         时间=1) # 修改已经存在的列
```

```python
df.assign(new=df.销售额*2,
         new_1 =df.时间,
         新_3=lambda x:x.new+10000,
         新_4=lambda x:x.dept+x['dept 1'],
         新_5=lambda x:x.dept+x['dept 1']+x.新_3.astype(str)) # 新增加的列 引用已经存在的多个列(包含过程中创建的列)
```

```python
df.assign(new=df.销售额*2,
         new_1 =df.时间,
         新_3=lambda x:x.new+10000,
         新_4=lambda x:x.dept+x['dept 1']) # 新增加的列 引用已经存在的多个列
```

```python
df.assign(new=df.销售额*2,
         new_1 =df.时间,
         新_3=lambda x:x.new+10000) # 新增加的列 可以引用前面的列 new
```

```python
df.assign(new=df.销售额*2,
         new_1 =df.时间) # 增加多列
```

```python
df.assign(new=df.销售额*2) # 简单增加一个新列
```
