##  95 pipe管道构建链式函数

### pipe 一般称为构建管道函数或链式函数。就是将业务数据流程中的函数“串成”流水线，看起来简洁

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['奖金']=np.random.randint(100,100000,size=(10, 1))
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[0,2]=np.nan # 修改一下
```

```python
df # 随机生成的数据
```

```python
df.groupby(['dept','dept1']).sum() # 作为对照
```

```python
def sum(data):
    return data.sum()
df.groupby(['dept','dept1']).pipe(sum) # 串自定义函数
```

```python
(df.groupby(['dept','dept1'])
.pipe(lambda x:x.mean())
.unstack()
 .round(1)
) # 偃师串函数的过程，这个写法看起来过程简明
```

```python
(df.groupby(['dept','dept1'])
.pipe(lambda x:x.mean())
.unstack()
)
```

```python
(df.groupby(['dept','dept1'])
.pipe(lambda x:x.mean())
)
```
