## 110 根据数据类型选择列

### select_dtypes ：根据数据类型选择列，返回指定数据类型的列
参数 include=None, exclude=None

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[1,3]=np.nan #  修改成nan
```

### 演示数据

```python
df 
```

```python
df.dtypes
```

```python
df.select_dtypes(include=['bool','object'],exclude=['float']) #同时使用，指定 包含的类型，和不包含的类型
```

```python
df.select_dtypes(include=['bool','object','float'],exclude=['bool','object','float']) #提示： 两个参数的 值 不能有重复部分
```

```python
df.select_dtypes(exclude=['bool','object','float']) # 返回bool 和object类型的列
```

```python
df.select_dtypes(include=['bool','object','float']) # 返回bool 和object类型的列
```

```python
df.select_dtypes(include=['bool','object']) # 返回bool 和object类型的列
```
