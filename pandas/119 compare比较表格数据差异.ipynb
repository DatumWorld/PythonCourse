## 119 compare比较表格数据差异

### compare：比较两个表格之间的数据差异，也可以比较列数据
参数 other, align_axis=1, keep_shape=False, keep_equal=False

```python
import numpy  as np 
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='3H')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df1=df.copy()
df1.iloc[1,3]=np.nan #  修改成nan
df1.iloc[1,4]=pd.Timestamp(2020,1,1,3,0,1) # 
df1.iloc[1,2]='1'
df1.iloc[2,1]='3'
df1.iloc[2,0]='部门'
```

```python
df.head()
```

```python
df1.head()
```

```python
df.compare(df1) # 将有差异的数据 的  位置 （行列） 差异值  表格方式 展现
```

```python
df.compare(df1,align_axis=0) # align_axis 控制结果的差异方式的排版显示  0  self other 沿着行方向竖排，1 则相反
```

```python
df.compare(df1,align_axis=0,keep_shape=1) # keep_shape 是否保持表格的形状不变（shape）
```

```python
df.compare(df1,align_axis=0,keep_shape=1,keep_equal=0) # keep_equal 相同的数据是否保持原来的值显示
```

```python
df.compare(df1,align_axis=0,keep_shape=1,keep_equal=1) # keep_equal 相同的数据是否保持原来的值显示
```
