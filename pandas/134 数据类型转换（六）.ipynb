## 134 数据类型转换（六）

### to_string 转换成字符串， to_period 转换成周期，  to_xarray 转换成 xarray，
### to_dict 转换成 字典， from_dict 字典转换成df，  read_json  读取json， to_json 保存成json

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[1,1]=np.nan #  修改成nan
```

### 演示数据

```python
df # 随机生成的数据
```

```python
pd.read_json('json.json') # 读取json文件
```

```python
df.to_json('json.json') # 转成json 格式字符串
```

```python
df.to_json() # 转成json 格式字符串
```

```python
df.from_dict(   df.to_dict()    ) # 读取字典
```

```python
df.to_dict() # 转换成字典
```
