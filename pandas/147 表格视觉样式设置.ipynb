## 147 表格视觉样式设置

### 表格视觉样式设置。目前不能保证设置的样式都能完整不变的输出保存到Excel
##### format 
##### hide_columns hide_index hide_columns_ hide_index_（pandas1.3支持）
更多使用请参考官方文档

```python

import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
df.iloc[1,1]=np.nan #  修改成nan
```

```python
df
```

```python
df.style.format( # 每列指定格式化输出的样式，lambda 函数
formatter={
    '销售额':lambda x:'{:.2f}'.format(x)
}
)
```

```python
df.style.format( # 每列指定格式化输出的样式，lambda 函数
formatter={
    '销售额':lambda x:'￥{:,.2f}'.format(x)
}
)
```

```python
df.style.format( # 每列指定格式化输出的样式，lambda 函数
formatter={
    '销售额':lambda x:'{:,.2f}'.format(x)
}
)
```

```python
df.style.format(
formatter={
    '销售额':'{:.2f}'
}
)
```

```python
df.style.hide_index_ # 属性。是否隐藏索引
```

```python
df.style.hide_index([0,3]) # 隐藏指定的索引行
```

```python
df.style.hide_index() # 隐藏索引
```

```python
df.style.hide_columns_ # 属性。是否隐藏列
```

```python
df.style.hide_columns(subset=['dept','name']) # 隐藏列
```

```python
df.style.hide_columns()
```
