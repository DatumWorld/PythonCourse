## 163 wide_to_long宽表调长表

###   wide_to_long：宽表调长表。与melt的区别：可以在列名上使用正则表达式拆解复合列名
##### 参数 df, stubnames, i, j, sep='', suffix='\\d+'

```python
import pandas as pd
import numpy  as np
```

```python
np.random.seed(123)
df = pd.DataFrame({"A1970" : {0 : "a", 1 : "b", 2 : "c"},
                   "A1980" : {0 : "d", 1 : "e", 2 : "f"},
                   "B1970" : {0 : 2.5, 1 : 1.2, 2 : .7},
                   "B1980" : {0 : 3.2, 1 : 1.3, 2 : .1},
                   "X"     : dict(zip(range(3), np.random.randn(3)))
                  })
df["id"] = df.index
df["id2"] = df.index
df
```

```python
pd.wide_to_long(df=df,stubnames=['A'],i='id',j='year') # df 数据源  ，stubnames 字符串跟，这里是A B ，i 索引列，j 是suffix拆后的列名称
```

```python
pd.wide_to_long(df=df,stubnames=['A','B'],i=['id','id2'],j='year') ## df 数据源  ，stubnames 字符串跟，这里是A B ，i 索引列，j 是suffix拆后的列名称
```

```python
np.random.seed(123)
df2 = pd.DataFrame({"A_1970" : {0 : "a", 1 : "b", 2 : "c"},
                   "A_1980" : {0 : "d", 1 : "e", 2 : "f"},
                   "B_1970" : {0 : 2.5, 1 : 1.2, 2 : .7},
                   "B_1980" : {0 : 3.2, 1 : 1.3, 2 : .1},
                   "X"     : dict(zip(range(3), np.random.randn(3)))
                  })
df2["id"] = df.index
df2["id2"] = df.index
df2
```

```python
pd.wide_to_long(df=df2,stubnames=['A','B'],i=['id','id2'],j='year',sep='_', suffix='\d+') ## sep 是stubnames 和 suffix 之间的分隔符
```
