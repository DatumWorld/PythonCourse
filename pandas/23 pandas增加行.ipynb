##  pandas增加行

### 数据行增加
#### append concat

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
```

```python
df.insert(0,'dept',['部门1']*5+['部门2']*5)
```

```python
df # 随机生成的数据
```

```python
df1=df # 创建一个df1，然后追加到df后面
```

```python
df.append(df1,ignore_index=1) # 将df1追加到df后边ignore_index=1 指定是否忽略索引行号
```

```python
pd.concat([df,df1],ignore_index=True) # 效果同上
```

```python
pd.concat([df,df1],axis=1) # concat的另一个使用方法，0横向追加行，1纵向追加列
```
