##  pandas表格合并

### 表格合并进阶
#### join merge concat

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
```

```python
df.insert(0,'dept',['部门1']*5+['部门2']*5)
```

```python
df # 随机生成的数据
```

```python
df1=df[['dept','name']]
df1['绩效系数']=np.random.randint(100,100000,size=(10, 1))
```

```python
df1  # 随机生成的数据
```

```python
df.set_index('dept').join(df1.set_index('dept'),lsuffix='_l') # 设置index后进行并表
```

```python
df.set_index(['dept','name']).join(df1.set_index(['dept','name'])) # 设置双列index后进行并表，这时候没有上述例子的重复现象
```

```python
pd.concat([df.set_index('dept'),df1.set_index('dept')]) # 设置index后合并表格。concat支持重复列名
```

```python
df.merge(df1,how='left',on=['dept']) #合并表格。dept为条件关联列
```

```python
df.merge(df1,how='left',on=['dept','name']) #合并表格。dept,name为条件关联列
```

```python
df2=df1.rename({'name':'name1'},axis=1)
```

```python
df2
```

```python
df.merge(df2,how='left',left_on=['dept','name'],right_on=['dept','name1']) #合并表格。当两边的关联列名不同时
```
