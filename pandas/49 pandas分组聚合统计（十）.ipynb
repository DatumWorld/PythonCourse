### 分组聚合统计
#### groupby  agg 使用 filter 根据条件过滤结果

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['奖金']=np.random.randint(100,100000,size=(10, 1))
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df # 随机生成的数据
```

```python
df.groupby(['dept','dept1'],as_index=0).agg({'销售额':['sum','mean']})
```

```python
df.groupby(['dept','dept1'],as_index=0).filter(lambda x :(x.销售额.sum()>56417) & (x.销售额.mean()>28208.500000) ) # 多个条件。使用 &的写法
```

```python
df.groupby(['dept','dept1'],as_index=0).filter(lambda x :x.销售额.sum()>56417 and x.销售额.mean()>28208.500000 ,dropna=0) # 多个条件
```

```python
df.groupby(['dept','dept1'],as_index=0).filter(lambda x :x.销售额.sum()>56417,dropna=0) # 将 销售额合计 大于 56417 的 组的 行过滤出来。 不符合的行以nan填充
```

```python
df.groupby(['dept','dept1'],as_index=0).filter(lambda x :x.销售额.sum()>56417) # 将 销售额合计 大于 56417 的 组的 行过滤出来
```
