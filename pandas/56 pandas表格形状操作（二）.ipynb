### stack unstack 将某个层级的列头调整到索引或相反

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['姓名'])
```

```python
df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['工龄']=np.random.randint(1,10,size=(10, 1))
df.insert(0,'部门',['部门1']*5+['部门2']*5)
df['绩效系数']=np.random.randint(7,10,size=(10, 1))/10
df['补贴']=np.random.randint(10,1000,size=(10, 1))
df['请假扣除']=np.random.randint(0,1000,size=(10, 1))
df['权重']=([0.2]+[.0])*5
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df # 随机生成的数据
```

```python
df1=df.groupby(['部门','dept1']).agg({'销售额':[np.sum,np.mean],'补贴':[np.sum,np.mean]})
```

```python
df1
```

```python
df2 = df1.stack()
```

```python
df2
```

```python
df1.unstack() # df1将小部门 调整到列头
```

```python
df2.unstack(level=1,fill_value='0') # 指定nan的填充
```

```python
df2.unstack(level=1) # 将某层级的索引调整到列头。默认最内层。层级可以指定
```

```python
df1.stack(level=0,dropna=1) #将指定层级的 列头调整到索引上。是否删除na
```

```python
df1.stack(level=0) #将指定层级的 列头调整到索引上
```
