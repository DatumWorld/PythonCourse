## 74 pandas虚拟指标哑变量

### get_dummies 将分类数据变量 转换为 虚拟/指标/哑变量。此处不解释其数学统计作用。
参数 data, prefix=None, prefix_sep='_', dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df.sort_values('销售额') # 随机生成的数据
```

```python
pd.get_dummies(data=pd.cut(df.销售额,[0.10000,50000,100000,110000]),prefix='销售额',prefix_sep='*',drop_first=1,dtype='float',dummy_na=1,sparse=True) 
# 和cut配合使用，切片后再用
```

```python
pd.get_dummies(data=df.销售额,prefix='销售额',prefix_sep='*',drop_first=1,dtype='float',dummy_na=1,sparse=True) 
# sparse 常规数组 还是 稀疏数组 编码，这里用不到
```

```python
df.iloc[1,3]=np.nan
df.iloc[1,3]=18952
```

```python
pd.get_dummies(data=df.销售额,prefix='销售额',prefix_sep='*',drop_first=1,dtype='float',dummy_na=1) 
# 如何处理nan。dummy_na=1 会对 nan 做出提示
# 注意修改后的变化
```

```python
pd.get_dummies(data=df.销售额,prefix='销售额',prefix_sep='*',drop_first=1,dtype='float') # dtype 指定结果的数据类型
```

```python
pd.get_dummies(data=df.销售额,prefix='销售额',prefix_sep='*',drop_first=1) # 余下面对比，注意少了最左边的第一个
```

```python
pd.get_dummies(data=df.销售额,prefix='销售额',prefix_sep='*') # prefix_sep 自定义 前缀分隔符
```

```python
pd.get_dummies(data=df.销售额,prefix='销售额') # prefix 自定义 前缀
```

```python
pd.get_dummies(data=df,columns=['dept','销售额']) # 可以指定具体使用哪些列
```
