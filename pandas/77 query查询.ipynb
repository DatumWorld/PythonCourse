### query 布尔表达式查询列

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

```python
df # 随机生成的数据
```

```python
dept_name='部门2'
df.query(''' dept==@dept_name and `dept 1` =='小部门1'   ''') # 使用@引入外部参数传递
```

```python
dept_name='部门1'
df.query(''' dept==@dept_name and `dept 1` =='小部门1' or not (10000<销售额<50000)  ''') # 使用@引入外部参数传递
```

```python
df[ (df.dept=='部门1') & (df['dept 1']=='小部门1') | ~((df.销售额>10000 ) & (df.销售额<50000))] # 普通逻辑查询。与下面等效
```

```python
df.query(''' dept=='部门1' and `dept 1` =='小部门1' or not (10000<销售额<50000)  ''') # 仅为展示 not 的用法
```

```python
df[ (df.dept=='部门1') & (df['dept 1']=='小部门1') | ((df.销售额>10000 ) & (df.销售额<50000))] # 普通逻辑查询。与下面等效
```

```python
df.query(''' dept=='部门1' and `dept 1` =='小部门1' or (10000<销售额<50000)  ''') # 仅为展示 or 的用法
```

```python
df.query(''' dept=='部门1' and `dept 1` =='小部门1' ''') # query 表达式查询。比较简便，像写查询表达式
# 注意dept 1 不是合法标识符。结果异常.  非合法列名的处理方法
```

```python
df.query(''' dept=='部门1' and 'dept 1' =='小部门1' ''') # query 表达式查询。比较简便，像写查询表达式
# 注意dept 1 不是合法标识符。结果异常
```

```python
df[ (df.dept=='部门1') & (df['dept 1']=='小部门1')] # 普通逻辑查询
```

```python
df.query(''' dept=='部门1' ''') # query 表达式查询。比较简便，像写查询表达式
```
