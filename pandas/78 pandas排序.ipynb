## 78 pandas排序

### sort_values  sort_index
参数 by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None

```python
import numpy  as np
import pandas as pd
from faker import Faker

fake = Faker('zh_CN')
df=pd.DataFrame(data=[ [fake.name()] for x in range(10) ],columns=['name'])

df['销售额']=np.random.randint(100,100000,size=(10, 1))
df['时间']=pd.date_range(start='2020-01-01',periods=10,freq='D')
df.insert(0,'dept',['部门1']*5+['部门2']*5)
df.insert(1,'dept 1',['小部门1']*2+['小部门2']*3+['小部门3']*1+['小部门4']*4)
```

### 演示数据

```python
df # 随机生成的数据
```

```python
df.iloc[1,1]=np.nan #  修改成nan
```

```python
df.set_index('时间').sort_index(ascending=False) # 按照索引排序
```

```python
df.dept.sort_values(ascending=False) # 也可以用于序列（列）
```

```python
df.sort_values(by=['dept 1'],ascending=[False],na_position='first',key= lambda x: x.str[-1],inplace=True) # inplace 参数 直接替换df而非返回修改后的副本
```

```python
df.sort_values(by=['dept 1'],ascending=[False],na_position='first',key= lambda x: x.str[-1]) # key 参数 自定义函数处理后再排序
```

```python
df.sort_values(by=['dept 1'],ascending=[False],na_position='first',kind='quciksort') # kind 参数 负责 排序方法。此处不多讲
```

```python
df.sort_values(by=['dept 1'],ascending=[False],na_position='first',ignore_index=1) # ignore_index 参数 负责 索引是否重新排列（自然数顺序）
```

```python
df.sort_values(by=['dept 1'],ascending=[False],na_position='first') # na_position, nan默认排在尾部； first 拍首位。注意区别
```

```python
df.sort_values(by=['dept','dept 1'],ascending=[False,False],na_position='first') # na_position, nan默认排在尾部； first 拍首位
```

```python
df.sort_values(by=['dept','dept 1'],ascending=[False,False]) # 按照dept 列 排序，ascending=False逆序。多列排序
```

```python
df.sort_values(by='dept',ascending=False) # 按照dept 列 排序，ascending=False逆序
```
